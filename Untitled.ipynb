### Step 0. Importing necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u
from astropy.cosmology import Planck18 as cosmo
import scipy.signal as signal
```

### Step 1. Loading the CSV data file

```python
file = np.loadtxt(fname='GW200129_065458_waveform.csv', skiprows=1, delimiter=',')
print(file.shape[:])
```

### Step 2. Plotting the Strain vs time waveform

```python
x = file[:, 0]
y = file[:, 1] 

#PLOTTING THE GRAPH.

plt.plot(x, y)
plt.xlabel('Time')
plt.ylabel('Strain')
plt.title('Strain vs time')
plt.tight_layout()
#plt.savefig('Strain vs time plot.png', dpi=300)
plt.show()
```

```python
# 1. Determine sampling properties from the time array
dt = x[1] - x[0]                     # time step between consecutive samples
fs = 1.0 / dt                        # sampling frequency in Hz

# 2. Design and apply a highpass Butterworth filter
f0 = 20.0                            # cutoff frequency of 20 Hz
order = 4                            # 4thorder filter gives a steep rolloff
sos = signal.butter(order, f0,
                      btype='highpass',
                      fs=fs,
                      output='sos')
y_hp = signal.sosfilt(sos, y)        # apply filter in secondorder sections form

# 3. Estimate the Power Spectral Density (PSD) with Welchs method
segment_length = int(fs * 4)         # use 4second data segments
f, Pxx = signal.welch(y_hp,
                      fs=fs,
                      nperseg=segment_length)

# 4. Plot the resulting amplitude spectral density (sqrt of PSD)
plt.figure(figsize=(6,4))
plt.loglog(f, np.sqrt(Pxx))         # loglog plot: frequency vs. ASD
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude Spectral Density [$1/\\sqrt{Hz}$]')
plt.title('Whitened-Strain ASD via Welch PSD')
plt.tight_layout()
plt.show()
```
